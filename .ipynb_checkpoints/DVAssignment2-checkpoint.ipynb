```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python



# STEP 1 — Load the US State shapefile using the .shp file

shapefile_path = "US_State.shp"
states_gdf = gpd.read_file(shapefile_path)

print("Shapefile loaded:")
print(states_gdf.head())



```

```python
# STEP 2 — Load the obesity CSV file

obesity_path_1 = "National_Obesity_By_State.csv"


obesity_df = pd.read_csv(obesity_path_1)



print("Obesity data loaded:")
print(obesity_df.columns)
print(obesity_df.head())


```

```python


# Remove missing values
obesity_df = obesity_df.dropna(subset=["NAME", "Obesity"])
obesity_df["NAME"] = obesity_df["NAME"].str.strip()

# View states
print("\nUnique states in obesity file:")
print(obesity_df["NAME"].unique())
```

```python

# Standardize both sides to lowercase
states_gdf.columns = states_gdf.columns.str.strip().str.lower()
obesity_df.columns = obesity_df.columns.str.strip().str.lower()

# Rename columns so they match shapefile
obesity_df.rename(columns={"name": "statename"}, inplace=True)

# STEP 4 — Merge shapefile with obesity data

merged = states_gdf.merge(obesity_df, on="statename", how="left")

print("Merged GeoDataFrame:")
print(merged.head())


```

```python

# STEP 5 — Find max & min obesity states

max_state = merged.loc[merged["obesity"].idxmax()]
min_state = merged.loc[merged["obesity"].idxmin()]

print("State with MAX obesity:", max_state["statename"])
print("State with MIN obesity:", min_state["statename"])


```

```python
# -----------------------------------------------------
# STEP 6 — Plot choropleth map
# -----------------------------------------------------
fig, ax = plt.subplots(figsize=(15, 10))

merged.plot(
    column="obesity",
    cmap="OrRd",
    linewidth=0.8,
    edgecolor="white",
    legend=True,
    ax=ax
)

# Highlight the max obesity state (red boundary)
gpd.GeoSeries(max_state.geometry).plot(
    ax=ax, edgecolor="green", linewidth=3
)

# Highlight the min obesity state (blue boundary)
gpd.GeoSeries(min_state.geometry).plot(
    ax=ax, edgecolor="black", linewidth=3
)

plt.title("Obesity Rates by US State\n(Red = Highest, Blue = Lowest)", fontsize=18)
plt.axis("off")
plt.show()
```

### Q3:Plot the distribution of School levels over states and indicate which highest, lowest of High and Elementary schools.

```python
schools = gpd.read_file("PublicSchools.shp")
schools.head()
```

```python
schools['LEVEL_'].unique()
```

```python
levels = schools[schools["LEVEL_"].isin(["ELEMENTARY", "HIGH"])]
level_counts = (
    levels.groupby(["STATE", "LEVEL_"])
          .size()
          .reset_index(name="count")
)

level_counts.head()
```

```python
level_pivot = level_counts.pivot(index="STATE", columns="LEVEL_", values="count")
```

```python
level_pivot = level_pivot.fillna(0)

level_pivot.head()
```

```python
sort_level = level_pivot.sort_values("ELEMENTARY", ascending=False)

x = np.arange(len(sort_level.index))
width = 0.4

plt.figure(figsize=(14, 8))

plt.bar(
    x - width/2,
    sort_level["ELEMENTARY"],
    width=width,
    label="Elementary",
    alpha=0.7,
    color="green" 
)

plt.bar(
    x + width/2,
    sort_level["HIGH"],
    width=width,
    label="High",
    alpha=0.7,
    color="orange" 
)

plt.xticks(x, sort_level.index, rotation=90)

plt.xlabel("State")
plt.ylabel("Number of Schools")
plt.title("Distribution of Elementary and High Schools Across States")
plt.legend(fontsize=14, borderpad=1.2, handlelength=2.5)

plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
highest_elem = level_pivot["ELEMENTARY"].idxmax()
lowest_elem = level_pivot["ELEMENTARY"].idxmin()

highest_high = level_pivot["HIGH"].idxmax()
lowest_high = level_pivot["HIGH"].idxmin()

highest_elem, lowest_elem, highest_high, lowest_high
```

### Plot the distribution of Schools’ FT_TEACHER over states and indicate which highest, lowest of High and Elementary schools.

```python
levels_teach = schools[schools["LEVEL_"].isin(["ELEMENTARY", "HIGH"])]

levels_teach.head()
```

```python
teacher_counts = (
    levels_teach.groupby(["STATE", "LEVEL_"])["FT_TEACHER"]
    .sum()
    .reset_index()
)

teacher_counts.head()
```

```python
teacher_pivot = teacher_counts.pivot(
    index="STATE", 
    columns="LEVEL_", 
    values="FT_TEACHER"
).fillna(0)

teacher_pivot.head()
```

```python
schools["FT_TEACHER"] = pd.to_numeric(schools["FT_TEACHER"], errors="coerce")
```

```python
schools.loc[schools["FT_TEACHER"] < 0, "FT_TEACHER"] = None
```

```python
levels_teach = schools[schools["LEVEL_"].isin(["ELEMENTARY", "HIGH"])]

teacher_counts = (
    levels_teach.groupby(["STATE", "LEVEL_"])["FT_TEACHER"]
    .sum()
    .reset_index()
)

teacher_pivot = teacher_counts.pivot(
    index="STATE",
    columns="LEVEL_",
    values="FT_TEACHER"
).fillna(0)

teacher_pivot.head()
```

```python
teacher_sorted = teacher_pivot.sort_values("ELEMENTARY", ascending=False)

x = np.arange(len(teacher_sorted.index))
width = 0.4

plt.figure(figsize=(14, 8))

plt.bar(
    x - width/2,
    teacher_sorted["ELEMENTARY"],
    width=width,
    label="Elementary FT Teachers",
    alpha=0.7,
    color='purple'
)

plt.bar(
    x + width/2,
    teacher_sorted["HIGH"],
    width=width,
    label="High FT Teachers",
    alpha=0.7,
    color='black'
)

plt.xticks(x, teacher_sorted.index, rotation=90)

plt.xlabel("State")
plt.ylabel("Number of Full-Time Teachers")
plt.title("Distribution of FT Teachers in Elementary and High Schools Across States")

plt.legend(fontsize=14, borderpad=1.2, handlelength=2.5)
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
highest_elem_teach = teacher_pivot["ELEMENTARY"].idxmax()
lowest_elem_teach = teacher_pivot["ELEMENTARY"].idxmin()

highest_high_teach = teacher_pivot["HIGH"].idxmax()
lowest_high_teach = teacher_pivot["HIGH"].idxmin()

highest_elem_teach, lowest_elem_teach, highest_high_teach, lowest_high_teach
```

### Q5: Plot the distribution of Schools’ ENROLLMENT and compare it with the plot in the Indicate those three regions which have lowest ENROLLMENT to FT_TEACHER ratio.

```python
schools["ENROLLMENT"].unique()[:20]
```

```python
schools["ENROLLMENT"] = pd.to_numeric(schools["ENROLLMENT"], errors="coerce")
schools.loc[schools["ENROLLMENT"] < 0, "ENROLLMENT"] = None
```

```python
levels_enroll = schools[schools["LEVEL_"].isin(["ELEMENTARY", "HIGH"])]
```

```python
enroll_counts = (
    levels_enroll.groupby("STATE")["ENROLLMENT"]
    .sum()
    .reset_index()
)

enroll_counts.head()
```

```python
enroll_sorted = enroll_counts.sort_values("ENROLLMENT", ascending=False)

plt.figure(figsize=(14,8))

plt.bar(enroll_sorted["STATE"], 
        enroll_sorted["ENROLLMENT"], 
        label="Total Enrollment", 
        alpha=0.7,
        color='red'
       )

plt.xticks(rotation=90)
plt.xlabel("State")
plt.ylabel("Total Enrollment")
plt.title("Distribution of Total Enrollment Across States")
plt.legend(fontsize=14, borderpad=1.2, handlelength=2.5)
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
ratio_df = teacher_pivot.copy()
ratio_df["ENROLLMENT"] = enroll_counts.set_index("STATE")["ENROLLMENT"]

ratio_df["Ratio"] = ratio_df["ENROLLMENT"] / (ratio_df["ELEMENTARY"] + ratio_df["HIGH"])

ratio_df.head()
```

```python


enroll_sorted = enroll_counts.set_index("STATE").sort_values("ENROLLMENT", ascending=False)

lowest_3 = ratio_df.nsmallest(3, "Ratio").index.tolist()

plt.figure(figsize=(14,8))

bars = plt.bar(enroll_sorted.index, enroll_sorted["ENROLLMENT"], 
               color="#4A90E2", alpha=0.7)

for bar, state in zip(bars, enroll_sorted.index):
    if state in lowest_3:
        bar.set_color("red")
        bar.set_alpha(0.9)

plt.xticks(rotation=90)
plt.xlabel("State")
plt.ylabel("Total Enrollment")
plt.title("Enrollment Across States with Lowest Enrollment/FT-Teacher Ratio Highlighted")

plt.legend(
    handles=[
        plt.Rectangle((0,0),1,1, color="#4A90E2", alpha=0.7),
        plt.Rectangle((0,0),1,1, color="red", alpha=0.9)
    ],
    labels=[
        "Total Enrollment",
        "Lowest Enrollment/FT-Teacher Ratio"
    ],
    fontsize=12
)

plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
